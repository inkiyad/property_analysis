```python
import requests
import pandas as pd
import json
```

## Collect Data

```python
url = "https://zillow56.p.rapidapi.com/search"

querystring = {"location":"Drexel Hill, PA"}
```

```python
## Example:
## Information can be collected from rapid api zillow https://rapidapi.com/s.mahmoud97/api/zillow56/
# headers = {
# 	"X-RapidAPI-Key": "xxxxxxxxxxxxxxxxxxxxxxxxxxxx",
# 	"X-RapidAPI-Host": "xxxxxxxxxxxxxxxxxx"
# }

with open('credentials.json') as f:
    headers = json.load(f)
```

```python
response = requests.get(url, headers=headers, params=querystring)
```

```python
df = pd.DataFrame(response.json()['results'])
df.head()
```

```python
df.to_csv('data/dec24.csv')
```

## Load Previous Data and Merge

```python
df1 = pd.read_csv('data/dec21.csv', index_col='Unnamed: 0')
df2 = pd.read_csv('data/dec24.csv', index_col='Unnamed: 0')

df = pd.concat([df1, df2])
```

```python
county_mill = 42.1414 / 1000
df['roi'] = 100 * (12 * df['rentZestimate'] - county_mill * df['taxAssessedValue']) / df['price']
```

```python
df.to_excel('data/upper_darby_drexel_hill_dec24_tax_rt.xlsx')
```
